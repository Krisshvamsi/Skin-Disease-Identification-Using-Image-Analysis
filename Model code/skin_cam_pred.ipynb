```python
from keras.models import load_model
from keras.preprocessing import image
import cv2
import numpy as np
```

```python
model = load_model("skin_proj.h5")
video=cv2.VideoCapture(0)
name=['cold sores', 'eczema', 'hives', 'melanoma', 'psoriasis']
```

```python
while(1):
    success,frame=video.read()
    cv2.imwrite('image.jpg',frame)
    img = image.load_img(r"image.jpg",target_size = (64,64))
    x=image.img_to_array(img)
    x = np.expand_dims(x,axis = 0)
    pred=model.predict_classes(x)
    p=pred[0]
    print(pred)
    cv2.putText(frame,"skin disease ="+str(name[p]),(120,170),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),1)
    cv2.imshow('image',frame)
    if cv2.waitKey(1) & 0xFF==ord('a'):
        break
video.release()
cv2.destroyAllWindows()
```
